```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from astropy.time import Time
from scipy.constants import c
from scipy.optimize import curve_fit

#%matplotlib qt

plt.rcParams['figure.figsize'] = [16, 8]
plt.rcParams['figure.facecolor'] = 'w'
```

```python
data = np.fromfile('/media/scott/fast/data/20201217/residuals.dat', sep = ' ').reshape((-1,4))
t_data = Time(data[:,0], format = 'mjd')
freq_data = data[:,1]
freq0_data = data[:,2]
res_data = data[:,3]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)


ax.plot(t_data.datetime,freq_data,'.',markersize=10, alpha=2, label = 'TLE Model')
ax.plot(t_data.datetime,freq0_data,'.',markersize=1, alpha=2, label = 'Data')

ax2 = plt.twinx()
ax2.plot(t_data.datetime, res_data*1000, '.',markersize=1, color = 'red', label = 'RMS')

lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc=9)

ax.set_title('CHANG-E 5 Orbiter Freqeuncy Data with Residuals')
ax.set_ylabel('Frequency (Hz)')
ax.set_xlabel('UTC time')

ax2.set_ylabel('Residuals (Hz)')
plt.show();
```
